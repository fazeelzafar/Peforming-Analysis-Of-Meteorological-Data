```python
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('weatherHistory.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['Daily Summary', 'Loud Cover'], axis=1, inplace=True)
df.head()
```

```python
df.columns[df.isnull().any()]
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.fillna('unknown', inplace=True)
df[df.isnull().any(axis=1)]
```

```python
df[['Formatted Date']]
```

```python
df['Formatted Date']=pd.to_datetime(df['Formatted Date'],utc=True)
df2 = df.set_index('Formatted Date')
df2 = df2.resample('M').mean()
df2.head()
```

```python
df2.info()
```

```python
import seaborn as sns
plt.figure(figsize=(12,8))
sns.relplot(x='Apparent Temperature (C)',y='Humidity',data=df2, kind='scatter').set(title='Relation b/w Apparent Temperature & Humidity')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(12,8))
sns.relplot(x='Apparent Temperature (C)',y='Humidity',data=df2, kind='scatter',  hue='Wind Speed (km/h)').set(title = 'Relation b/w Apparent Temperature wrt to Wind Speed')
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.relplot(x='Apparent Temperature (C)',y='Humidity', hue = 'Summary', data = df, col = 'Summary', col_wrap = 4)
plt.show()
```
